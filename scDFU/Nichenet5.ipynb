```r
Idents(seuratObj) <- "group"

sender_cells <- subset(seuratObj, cell_type %in% c("Mac_VI", "VSMC_FB", "VSMC_Con"))

DE_sender <- FindMarkers(sender_cells, ident.1 = "DFU", ident.2 = "control", logfc.threshold = 0)

DE_sender$gene <- rownames(DE_sender)

ligands <- lr_network$from %>% unique()

ligand_DE_table <- DE_sender %>% filter(gene %in% ligands)

head(ligand_DE_table[, c("gene", "avg_log2FC", "p_val_adj")])
```

```r
celltypes <- c("Mac_VI", "VSMC_FB", "VSMC_Con")
ligands <- c("PGF","VEGFA","IGF1","CXCL12","AGT")

results_list <- list()

for (ct in celltypes) {
  message("Processing cell type: ", ct)
  

  cells <- WhichCells(seuratObj, idents = ct)
  
  if (length(cells) < 2) next  
  
  subset_obj <- subset(seuratObj, cells = cells)
  
  
  if (!"group" %in% colnames(subset_obj@meta.data)) {
    stop("Seurat meta data 中没有 'group' 列")
  }
  

  markers <- FindMarkers(
    object = subset_obj,
    ident.1 = "DFU",
    ident.2 = "control",
    group.by = "group",
    features = ligands,
    logfc.threshold = 0  
  )
  
  markers <- markers %>%
    rownames_to_column("gene") %>%
    mutate(celltype = ct) %>%
    select(celltype, gene, avg_log2FC = avg_log2FC, p_val_adj = p_val_adj)
  
  results_list[[ct]] <- markers
}

plot_data <- bind_rows(results_list)
```

```r
print(plot_data)
write.csv(plot_data,
          file = "DE_DFU_vs_control_by_celltype.csv",
          row.names = FALSE)
```

```r
selected_ligands <- c( "AGT", "CXCL12","IGF1", "VEGFA","PGF")

ligand_receptor_df <- nichenet_output$ligand_receptor_df %>%
  filter(ligand %in% selected_ligands) %>%
  select(ligand, receptor, weight) %>%
  rename(Ligand = ligand, Receptor = receptor, lr_weight = weight)


ligand_target_df <- nichenet_output$ligand_target_df %>%
  filter(ligand %in% selected_ligands) %>%
  group_by(ligand) %>%
  top_n(8, weight) %>%  
  ungroup() %>%
  select(ligand, target, weight) %>%
  rename(Ligand = ligand, Target = target, lt_weight = weight)

alluvial_data <- ligand_receptor_df %>%
  left_join(ligand_target_df, by = "Ligand") %>%
  mutate(weight = lr_weight * lt_weight * 50) %>%  
  select(Ligand, Receptor, Target, weight) %>%
  filter(!is.na(Target)) %>%
  mutate(
    Ligand = factor(Ligand, levels = selected_ligands),
    Receptor = factor(Receptor),
    Target = factor(Target)
  )

cat("\n=== 桑基图统计信息 ===\n")
cat("配体数量:", length(unique(alluvial_data$Ligand)), "\n")
cat("受体数量:", length(unique(alluvial_data$Receptor)), "\n")
cat("靶基因数量:", length(unique(alluvial_data$Target)), "\n")
cat("总连接数:", nrow(alluvial_data), "\n\n")

for(lig in selected_ligands) {
  if(lig %in% alluvial_data$Ligand) {
    n_receptors <- alluvial_data %>% filter(Ligand == lig) %>% pull(Receptor) %>% unique() %>% length()
    n_targets <- alluvial_data %>% filter(Ligand == lig) %>% pull(Target) %>% unique() %>% length()
    cat(sprintf("%s: %d 个受体, %d 个靶基因\n", lig, n_receptors, n_targets))
  }
}

p <- ggplot(alluvial_data,
       aes(y = weight,
           axis1 = Ligand,
           axis2 = Receptor,
           axis3 = Target)) +
  geom_alluvium(aes(fill = Ligand), 
                width = 1/4,  
                alpha = 0.6,
                curve_type = "cubic",
                decreasing = NA) +  
  geom_stratum(width = 1/4,  
               fill = "white", 
               color = "gray40",
               alpha = 0.9,
               linewidth = 0.5,
               decreasing = NA) +  
  geom_text(stat = "stratum", 
            aes(label = after_stat(stratum)),
            size = 3.2,
            fontface = "bold",
            color = "gray20",
            lineheight = 0.8,
            decreasing = NA) +  
  scale_x_discrete(limits = c("Ligand", "Receptor", "Target"),
                   expand = c(0.05, 0.05),
                   labels = c("Ligand", "Receptor", "Target")) +
  scale_fill_manual(
    values = c(
      "PGF" = "#E78AC3",
      "VEGFA" = "#A6D854",
      "IGF1" = "#8DA0CB",
      "CXCL12" = "#FC8D62",
      "AGT" = "#66C2A5"
    ),
    breaks = selected_ligands 
  ) +
  coord_cartesian(clip = "off") + 
  theme_minimal() +
  theme(
    legend.position = "bottom",
    legend.title = element_text(size = 13, face = "bold"),
    legend.text = element_text(size = 11),
    legend.key.size = unit(1, "cm"),
    axis.text.x = element_text(size = 14, face = "bold", color = "gray30"),
    axis.text.y = element_blank(),
    axis.title = element_blank(),
    panel.grid = element_blank(),
    panel.background = element_rect(fill = "white", color = NA),
    plot.background = element_rect(fill = "white", color = NA),
    plot.title = element_text(hjust = 0.5, size = 16, face = "bold", margin = margin(b = 20)),
    plot.margin = margin(20, 20, 20, 20)
  ) +
  labs(title = "Ligand-Receptor-Target Interactions",
       fill = "Ligand")
print(p)
```
